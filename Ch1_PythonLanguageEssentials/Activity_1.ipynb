## Activity \#1
#### Problem Description
초등학생들이 여러 자리 수의 덧셈을 배울 때는 한 번에 한 자리씩 오른쪽에서 왼쪽으로 계산하도록 배운다. 그런데 그 자리 숫자의 합이 10을 넘어갈 때 그 윗자리 숫자에 1을 더해주는 것을 배울 때 많은 학생들이 힘들어한다. 일련의 덧셈 문제가 주어졌을 때 자리를 올리는 횟수를 세어서 선생님들이 학생들을 가르치는 데 도움을 줄 수 있는 프로그램을 만들어야 한다.

#### Input:
>  각 행에는 열 자리 미만의 부호가 없는 정수가 두 개씩 입력된다. 마지막 줄에는 '0 0'이 입력된다.

#### Output:
>  마지막 줄을 제외한 각 줄에 대해 주어진 두 수를 더할 때 자리를 올려야 하는 횟수를 계산한 다음, 아래에 주어진 형식대로 결과를 출력한다.

#### Sample Input:
> 123 456

> 555 555

> 123 594

> 0 0

#### Sample output:
> No carry operation.

> 3 carry operations.

> 1 carry operation.

```python
# 두 숫자의 덧셈 과정에서 발생하는 carry를 count하는 함수
# input : 더하고자 하는 두 숫자(문자열 타입으로 인자 전달)
# ouput : carry 발생 횟수
def countCarryOperation(firstNum, secondNum):
    reversedFirstNum = []
    reversedSecondNum = []
    
    # 숫자를 뒤짚고(firstNum[::-1]) 정수형의 리스트(reversedFirstNum)로 변환
    for num in firstNum[::-1]:
        reversedFirstNum.append(int(num))
    
    # 숫자를 뒤짚고(secondNum[::-1]) 정수형의 리스트(reversedSecondNum)로 변환
    for num in secondNum[::-1]:
        reversedSecondNum.append(int(num))

    # carry가 발생하면 1, 발생하지 않으면 0
    # carryCount는 덧셈 과정에서 발생하는 carry의 누적 횟수
    carry = 0
    carryCount = 0
    
    # 두 숫자 중 긴 숫자에 맞춰 반복
    for i in range(0, max(len(reversedFirstNum), len(reversedSecondNum))):
        # 첫번째 숫자의 길이가 긴 숫자보다 짧을 경우 0으로 대체
        if(i < len(reversedFirstNum)):
            firstNumDigit = reversedFirstNum[i]
        else:
            firstNumDigit = 0
        
        # 두번째 숫자의 길이가 긴 숫자보다 짧을 경우 0으로 대체
        if(i < len(reversedSecondNum)):
            secondNumDigit = reversedSecondNum[i]
        else:
            secondNumDigit = 0
        
        # 각 자리수별 덧셈
        if((firstNumDigit + secondNumDigit + carry) >= 10):
            carry = 1
            carryCount += 1
        else:
            carry = 0
    
    return carryCount


#무한 루프
while True:
    # 입력받으면 기본 자료형은 문자열
    firstNum = input("첫번째 숫자를 입력하세요 : ")
    secondNum = input("두번째 숫자를 입력하세요 : ")
    
    # 종료 조건
    if(firstNum == '0' and secondNum == '0'): break
    
    # carry 발생 횟수를 count하기 위한 변수
    carryCount = 0
    
    # 문제 해결을 위한 함수 실행
    carryCount = countCarryOperation(firstNum, secondNum)
    
    # 출력
    if(carryCount == 0):
        print("No carry operation.")
    else:
        print("%d carry operation." % carryCount)
```
